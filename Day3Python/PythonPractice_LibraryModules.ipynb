```python
#make numpy array consisting of #1-10. Call it 'xarray'. 
#Make numpy array that consists of these # doubled(2,4,...20). 
#Call it yarray
xarray = np.arange(1,11,1)
xarray
```

```python
yarray = np.arange(2,22,2)
yarray
```

```python
zarray = np.power(xarray,3)
zarray
```

```python
zzarray = xarray **3
zzarray
```

```python
#Lets make out 1st plot
plt.plot(xarray, zarray,color='b')
plt.title('Super best graph ever!')
```

```python
plt.plot(yarray, xarray,color='r')
```

```python
plt.semilogy(zarray, np.log10(zarray))
plt.title('Super best graph ever!')
plt.grid(b=True)
```

```python
plt.plot(yarray, xarray,color='r')
plt.yscale('log')
```
